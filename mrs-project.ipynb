**PRE-PROCESSING ON THE DATA**

```python
#pre-processing on the data set
movies.isnull().sum()           #3 NAN values found in the column 'overview'
```

```python
movies.dropna(inplace = True)         #dropping all the NAN rows from the dataset since there were only very few missing values
```

```python
movies.duplicated().sum()     #No duplicated values found
```

```python
movies['genres'][0]
```

```python
#exracting name from dict in column genres
#importing ast lib to convert string of list to list
import ast
def name(data):
  gen = []
  for i in ast.literal_eval(data):
    gen.append(i['name'])
  return gen
```

```python
movies['genres'] = movies['genres'].apply(name)
```

```python
movies['keywords'][0]
```

```python
#applying the same function to column keywords
movies['keywords'] = movies['keywords'].apply(name)
```

```python
movies['cast'][0]
```

```python
#creating function for cast column for extracting first 4 cast name
def name_cast(data):
  gen = []
  counter = 0
  for i in ast.literal_eval(data):
      if counter != 4:
        gen.append(i['name'])
        counter+=1
      else:
        break

  return gen
```

```python
movies['cast'] = movies['cast'].apply(name_cast)   #applying it to column cast
```

```python
movies['crew'][0]
```

```python
#extracting name of the director only from column crew
def name_crew(data):
  gen = []
  for i in ast.literal_eval(data):
    if i['job'] == 'Director':
      gen.append(i['name'])
      break
  return gen
```

```python
movies['crew']=movies['crew'].apply(name_crew)   #applying the function and saving it to the dataset
```

```python
movies['overview'][0]
```

```python
#overview is a str and I will convert it to a list so that we can concat it to the other columns
movies['overview'] = movies['overview'].apply(lambda x : x.split())
```

```python
#removing spaces between words within the list
movies['genres'] = movies['genres'].apply(lambda x : [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x : [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x : [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
#joining columns overview,genres,keywords,cast,crew to a single column named tags
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['id','title','tags']]
```

```python
new_df.head()
```

```python
#coverting tags columns to str
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df.head()
```

```python
#converting everything in lower form
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df['tags'][0]
```

**Vectorization**

```python
!pip install nltk
import nltk
nltk.download('wordnet')
```

```python
#lemmatization of the column tags to reduce dimensionality
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
def lem(text):
  Y=[]
  for i in text.split():
    Y.append(lemmatizer.lemmatize(text))
  return " ".join(Y)
```

```python
new_df['tags'] = new_df['tags'].apply(lem)
```

```python
#vectorizing the text using bag-of-word techquie
#importing lib for vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words = 'english')   #creating a object named cv which will be our model to vectorize the text
                                                              #it will also exclude the english stop words
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

**Distance calculation using cosine similarity**

```python
#we will calculate the distance between vector of all moives to check the similarities between movies
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors) #its the distance between each movie
```

```python
similarity.shape
```

```python
similarity
```

**Main function**

```python
def recommend (movie):
  movie_index = new_df['title'][new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse = True, key= lambda x:x[1])[1:6]

  for i in movies_list:

    print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman')
```
